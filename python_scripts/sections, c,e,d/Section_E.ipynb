```python
import numpy as np
from scipy.stats import zscore

# Standard deviations of consumption for each consumer (calculated by the Tableau)
std_devs = np.array([592.5, 102, 112.7, 866.7])

# Calculate z-scores for each standard deviation
z_scores = zscore(std_devs)

# Map z-scores to a 1-10 scale
scores = 10 * (z_scores + 2) / 4 

# Round scores to 2 decimal places
rounded_scores = np.round(scores, 2)

print("The Scores are:")
# Print message with scores for each consumer
for i, score in enumerate(rounded_scores):
    print(f"Consumer {i+1}: {score}")
```
