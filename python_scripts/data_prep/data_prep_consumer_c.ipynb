# Merge the cols to get new cols that presents intervals of half hour

```python
# make a copy of the data frame
copy_df = df

# Loop over the odd columns in reverse order so that I can merge the columns 
# and drop the unnecessary in the same iteration
for i in range(len(df.columns)-2, 0, -2):
  copy_df[copy_df.columns[i]] += copy_df[copy_df.columns[i+1]]
  copy_df.drop(copy_df.columns[i+1], axis=1, inplace=True)

copy_df
```

# Arrange the data frame in 3 cols of : date, hour, kwh

```python
# melt the dataframe, keeping 'index' and 'date' as id_vars
melted_df = df.melt(id_vars=['date'], var_name='hour', value_name='kWh_amount')
# extract the hour value from the column name
melted_df['hour'] = melted_df['hour'].apply(lambda x: x)
# sort the dataframe by date and hour
melted_df = melted_df.sort_values(['date', 'hour'])
# reset the index
melted_df = melted_df.reset_index(drop=True)
melted_df
```

# Outliers Detection

```python
fig = go.Figure()
fig.add_trace(go.Box(name='kWh amount', y=df.kWh_amount, boxmean='sd'))

fig.update_traces(
    marker=dict(
        color='rgb(8,81,156)',
        outliercolor='rgba(219, 64, 82, 0.6)',
        line=dict(
            outliercolor='rgba(219, 64, 82, 0.6)',
            outlierwidth=2)),
    line_color='rgb(8,81,156)'
)
fig.show()
```

```python
def get_outliers(df, series):
  q1 = series.quantile(0.25)
  q3 = series.quantile(0.75)

  if q1*q3 == 0:
    iqr = abs(2*(q1+q3))
    toprange = iqr
    botrange = -toprange
  else:
    iqr = q3-q1
    toprange = q3 + iqr * 1.5
    botrange = q1 - iqr * 1.5

  outliers_top=df[series > toprange]
  outliers_bot= df[series < botrange]
  outliers = pd.concat([outliers_bot, outliers_top], axis=0)

  return (botrange, toprange, outliers)
```

```python
botrange, toprange, outliers = get_outliers(df, df.kWh_amount)
print(toprange)
print(botrange)

outliers
```
